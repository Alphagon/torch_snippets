... or use track2 to directly track a loop

```python
N = 100
info = None

for i in (tracker := track2(range(N), total=N)):
    time.sleep(0.1)
    info = f"My Info: {i*3.122:.2f}"
    if i == N // 2:
        print()
    if i >= N // 2:
        tracker.send(info)
```
