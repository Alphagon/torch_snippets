Suppose you have a file called `config.ini` like so...

```ini
[project]
version = 0.0.1
name = mnist
root = /home/me/projects/${project.name}

[project.data]
source = https://files.fast.ai/data/examples/mnist_tiny.tgz
root = ${project.root}/data/

[test]
@test_func = hello
```

You can load it up as an `AttrDict`

````python
from torch_snippets.registry import parse
config = parse("/tmp/config.ini")
assert config.project.version == "0.0.1"
assert config.project.root == '/home/me/projects/mnist'
config.pretty()
````

Notice, how the `${}` variables got resolved.

You can also register python functions/classes/objects to strings by running

````python
from torch_snippets.registry import registry

registry.create("test_func")

@registry.test_func.register("hello")
def test():
    return lambda: "hello world"
````

... and resolve them on parse

````python
from torch_snippets.registry import parse_and_resolve
config = parse_and_resolve("/tmp/config.ini")
assert config.test() == "hello world"
````
