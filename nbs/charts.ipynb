```python
from sklearn.datasets import make_moons

x, y = make_moons(1000, noise=0.1)
df = pd.DataFrame({'x1': x[:,0], 'x2': x[:,1], 'y': y})

Chart(df).mark_circle().encode(
    x='x1:Q',
    y='x2:Q',
    color='y:N'
).interactive()
```
